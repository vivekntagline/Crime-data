```python
# Replacing Null Values

df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna('No Weapon Used')
df['Weapon Desc'] = df['Weapon Desc'].fillna('No Weapon Used')

for col in ["Crm Cd 1", "Crm Cd 2", "Crm Cd 3", "Crm Cd 4"]:
    if col in df.columns:
        df[col] = df[col].fillna("No Additional Crime")

for col in ["Vict Sex", "Vict Descent", "Status"]:
    if col in df.columns:
        df[col] = df[col].fillna("Other")

for col in ["Mocodes", "Premis Cd", "Premis Desc", "Cross Street"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

```python
# Convert Time format from xxxx to xx:xx
df["TIME OCC"] = df["TIME OCC"].astype(str).str.zfill(4)  # Ensure all are 4 digits
df["TIME OCC"] = df["TIME OCC"].str[:2] + ":" + df["TIME OCC"].str[2:]  # Add colon

# every row in the "Date Rptd" and "DATE OCC" columns has a timestamp (12:00:00 AM), 
# so remove the time portion and keep only the date make sense.

df["Date Rptd"] = pd.to_datetime(df["Date Rptd"]).dt.date
df["DATE OCC"] = pd.to_datetime(df["DATE OCC"]).dt.date
```
